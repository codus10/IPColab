```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Ch01/다운로드.jpg')
print(img.shape)
glmg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(glmg1.shape)
plt.imshow(glmg1, cmap='gray')
```

```python
def saturation(val):
  if val > 255:
    return 255
  elif val < 0:
    return 0
  else:
   return val
```

```python
gPoint = np.zeros((glmg1.shape[0], glmg1.shape[1]), dtype=np.ubyte)
```

```python
th = 80

for h in range(glmg1.shape[0]):
  for w in range(glmg1.shape[1]):
    if glmg1[h,w] > th:
      gPoint[h,w] = 255
    else:
      gPoint[h,w] = 0


plt.imshow(gPoint, cmap='gray')
```

```python
num0fClip = 3
modVal = 256/num0fClip

for h in range(glmg1.shape[0]):
  for w in range(glmg1.shape[1]):
    piontVal = np.int32(glmg1[h,w]) % modVal
    gPoint[h,w] = saturation(piontVal * num0fClip)


plt.imshow(gPoint, cmap='gray', vmin=0, vmax=255)
```
